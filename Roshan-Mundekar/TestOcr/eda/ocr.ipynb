```python
import os
import cv2
import pytesseract

texts = []

# Define paths
samples_dir = os.path.abspath(os.path.join(os.getcwd(), "..", "samples"))

# Loop through image files
for image_name in os.listdir(samples_dir):
    image_path = os.path.join(samples_dir, image_name)

    # Read image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if image is None:
        print(f"❌ Failed to read image: {image_name}")
        continue

    # Run Tesseract OCR
    text = pytesseract.image_to_string(image)
    texts.append((image_name, text))

print("✅ OCR complete. Extracted text from all images.")
```
