```python
import pandas as pd

# Path to the Excel file
file_path = "a.xlsx"

# Load the Excel file
excel_data = pd.ExcelFile(file_path)

# Define the sheet name
sheet_name = excel_data.sheet_names[10]

# Parse the selected sheet
df = excel_data.parse(sheet_name)

# output_csv = "output_sheet.csv"
# df.to_csv(output_csv, index=False)


# print(df.head(50))
# Initialize the result DataFrame with required columns
ansDF = pd.DataFrame(columns=['SKU', 'Sheet', 'Measure', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Find the column where row 7 has "Pack Link"
pack_link_column = df.columns[(df.iloc[7] == 'Pack Link').to_numpy()]
print(f"pack_link_column ::: {pack_link_column}")

if not pack_link_column.empty:
    column_name = pack_link_column[0]
    entries_to_append = df.loc[8:, column_name].dropna().tolist()
    
    selected_row = df.iloc[6]
    for row_index, measure in enumerate(selected_row.dropna()):
        
        # Prepare data for `SKU` column in `ansDF`
        # for index, measure1 in selected_row.items():
        #     if pd.notna(measure1):
        #         # print(f"{index}: {measure1}")
        #         for indexcolumn, month_presnt in df.iloc[7,index+2:index+14].items():
        #           print(f"{indexcolumn}: {month_presnt}")

        start_col_index = row_index + 2  # Adjust based on your DataFrame structure
        month_columns = df.columns[28:35]
        print(f"{month_columns}: {month_columns}")
        # Extract month values
        month_values = df.iloc[8:, 28:35].fillna(0).values.tolist()
        print(f"{month_values}: {month_values}")
        for sku, month_vals in zip(entries_to_append, month_values):
            new_row = {
                'SKU': sku,
                'Sheet': sheet_name,
                'Measure': measure,
                **dict(zip(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], month_vals))
            }
            ansDF = pd.concat([ansDF, pd.DataFrame([new_row])], ignore_index=True)
    
    print(ansDF.head(10))  # Display the top rows of the result DataFrame
    # output_file = "output.csv"

    # Save the DataFrame as a CSV file
    # ansDF.to_csv(output_file, index=False)

    # print(f"CSV file saved successfully at {output_file}.")
else:
    print("No column found with 'Pack Link' in row 7.")
```
