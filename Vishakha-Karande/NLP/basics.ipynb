What is tokenization?

Breaking text into small units called tokens

Usually:

Words

Sometimes sub-words

Sometimes characters

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# Split cleaned text into tokens
tokens = cleaned_text.split()
print(tokens)
```

Why tokenization is critical

Because:

ML models work on lists, not sentences

Vocabulary is built from tokens

Embeddings are created per token

LLMs predict next token, not next sentence

üëâ Tokenization is the entry gate to LLMs.
Stopwords (WHY remove them?)

Stopwords = very common words:

is, the, a, an, to, in, of


They:

Don‚Äôt add meaning

Increase noise

Increase computation



```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
filtered_tokens=[word for word in tokens if word not in stop_words]
print(filtered_tokens)
```

```python
import spacy
spacy.cli.download("en_core_web_sm")
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("I LOVE learning!!! Machine Learning is AMAZING.")

tokens_spacy = []

for token in doc:
    if token.is_alpha and not token.is_stop:
        tokens_spacy.append(token.text.lower())

print(tokens_spacy)
```
