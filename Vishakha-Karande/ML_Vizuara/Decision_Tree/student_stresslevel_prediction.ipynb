```python
data.drop(columns=["Student_ID"],axis=1) #dropping Student_ID column as it is not useful for prediction
```

```python
from sklearn.preprocessing import LabelEncoder #Importing LabelEncoder from sklearn
label_encoder=LabelEncoder() #Creating object of LabelEncoder
data["Stress_Level"]=label_encoder.fit_transform(data["Stress_Level"]) #Encoding target column
```

```python
data.head()
```

```python
x=data.drop(columns=["Stress_Level"],axis=1) #features
y=data["Stress_Level"] #target variable
from sklearn.model_selection import train_test_split #importing train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) #splitting data into training and testing sets
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="gini",max_depth=3) #Creating Decision Tree Classifier model
model.fit(x_train,y_train) #Training the model  
```

```python
y_predict=model.predict(x_test) #Predicting on test data
```

```python
result = label_encoder.inverse_transform(y_predict)
print("Prediction:", result[0:10])  #display first 10 predictions
```

```python
print("Y_test:", label_encoder.inverse_transform(y_test.values[0:10]))  #display first 10 actual values
```

```python
#compare the y_predict and y_test to evaluate model performance
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy=accuracy_score(y_test,y_predict)   #calculating accuracy
confusion=confusion_matrix(y_test,y_predict) #confusion matrix
report=classification_report(y_test,y_predict) #classification report
print("Accuracy:",accuracy)
print("Confusion Matrix:\n",confusion)
print("Classification Report:\n",report)
accuracy #display accu
```
