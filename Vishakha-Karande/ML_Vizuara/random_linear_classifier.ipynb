```python
np.random.seed(42)
```

```python
dog_whisker = np.random.normal(5, 1, 10)
dog_ear = np.random.normal(8, 1, 10)
```

```python
def random_linear_classifier(dog_data, cat_data, K, D):
    best_error = float('inf')
    for i in range(K):
        theta = np.random.normal(0, 1, D)
        theta0 = np.random.normal(0, 1)
        current_error = compute_error(theta, theta0, dog_data, cat_data)
        if current_error < best_error:
            best_error = current_error
            best_theta, best_theta0 = theta, theta0
    return best_theta, best_theta0
```

```python
def compute_error(theta, theta0, dog_data, cat_data):
    error = 0
    # Check dogs (should be positive)
    for x in dog_data:
        if np.dot(theta, x) + theta0 <= 0:
            error += 1
    # Check cats (should be negative)
    for x in cat_data:
        if np.dot(theta, x) + theta0 >= 0:
            error += 1
    return error
```

```python
# Define dogs data using existing variables
dogs = np.column_stack((dog_whisker, dog_ear))

# Generate cats data with different means for classification
cat_whisker = np.random.normal(6, 1, 10)
cat_ear = np.random.normal(7, 1, 10)
cats = np.column_stack((cat_whisker, cat_ear))

best_theta, best_theta0 = random_linear_classifier(dogs, cats, 100, 2)
print(best_theta, best_theta0)
```
