<!-- 1. Supervised Learning -->
This is the most common type of ML you will encounter.
• The Core Concept: You act like a teacher with flashcards. You give the computer Input-Output Pairs (Questions and Answers).
• The Definition: The goal is to learn the relationship (mapping) between Input (X) and Output (Y) so that when the computer sees a new X, it can predict the correct Y.
• Real-World Examples:
    ◦ Healthcare (Heart Attack Prediction):
        ▪ Input (X): Blood pressure, glucose levels.
        ▪ Output (Y): Will the patient have a heart attack? (Yes/No).
    ◦ Image Recognition (Cats vs. Dogs):
        ▪ Input: Images.
        ▪ Output: Label "Cat" or "Dog".
    ◦ Binary Classification: A specific type of supervised learning where there are only two possible outputs (e.g., Malignant vs. Benign tumor).
The Practical Demo (Teachable Machine): In the lecture, we used Google's "Teachable Machine" to demonstrate this.
1. Training: We uploaded 10 images of cats and 10 images of dogs. We explicitly told the model "These are cats" and "These are dogs" (Supervision).
2. Testing: We uploaded a new, unseen image of a dog (side view).
3. Result: The model correctly identified it as a dog because it learned the patterns from the labeled data we provided.

2. Unsupervised Learning
This is where we stop "teaching" and let the machine explore.
• The Core Concept: There are NO labels and NO correct answers. We just dump data into the machine and ask it to find structure.
• The Definition: We are given a dataset and expected to find patterns or structure within it, without any input-output pairs.
• Real-World Examples:
    ◦ Amazon Recommendations: Amazon doesn't have a "correct answer" for what you should buy. Instead, it looks at millions of users and finds clusters: "Customers who bought this microphone also bought this tripod." It groups similar users together based on purchasing patterns.
    ◦ The Iris Flower Dataset: Imagine a botanist finds a field of flowers but doesn't know their species. They measure petal length and width.
        ▪ The Task: The algorithm (K-Means Clustering) looks at the data and says, "Hey, these flowers naturally group into three distinct piles based on size." It discovers the species on its own.

3. Reinforcement Learning (RL)
This is the most complex type, often used in robotics and gaming.
• The Core Concept: Training a pet or playing a video game. It’s about Action and Reward.
• The Definition: An Agent interacts with an Environment. It takes an action and receives either a Reward (good) or a Punishment (bad). The goal is to maximize the total reward over time.
• How it works:
    1. Observe: The agent sees the current state.
    2. Action: The agent makes a move.
    3. Reward: The environment tells the agent if the move was good (points) or bad (game over).
    4. Update: The agent learns from this feedback to make better moves next time.
• Real-World Examples:
    ◦ AlphaGo & Chess: The AI plays millions of games against itself. If it wins, the moves it made are "reinforced" (rewarded). If it loses, it learns to avoid those moves. This is how AI beat the world chess champion Gary Kasparov.
