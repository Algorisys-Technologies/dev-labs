1. The Hierarchy (The Venn Diagram)
Imagine a set of Russian nesting dolls.
• Artificial Intelligence (AI): The biggest doll. This is any machine that mimics human intelligence.
    ◦ Example: A Rule-Based Chatbot. If you program a bot with 100 fixed questions and answers (If user says "Hi", print "Hello"), that is AI, but it is not ML because it isn't learning; it's just following rules.
• Machine Learning (ML): A subset inside AI. This is where the machine uses data to learn and improve predictions without being explicitly programmed for every rule.
    ◦ Example: Netflix Recommendations. It learns from your watch history to suggest new movies. It improves over time.
• Deep Learning (DL): A subset inside ML. This deals specifically with Neural Networks (algorithms inspired by the human brain).
    ◦ Example: Self-Driving Cars or Image Recognition.
• Data Science: This is a separate circle that intersects with all of them. It involves using statistics and visualization to find patterns in data.
2. Why is ML "Hyped" Now? (The History)
ML isn't new. The term was coined in the 1950s. So why is it everywhere now?
• 1956-1974 (Golden Years): Early excitement, robots like "Shakey".
• 1974-1993 (AI Winter): Funding dried up. People lost faith because the algorithms couldn't solve real problems.
• 1993-2011 (The Boom): The internet happened.
• Why Now? Three things exploded simultaneously:
    1. Data: We have massive amounts of data (Big Data).
    2. Compute: We have powerful GPUs/CPUs to process that data.
    3. Algorithms: We finally developed the math to make deep learning work.
