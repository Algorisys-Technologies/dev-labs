Goal: Set up the professional environment required for Machine Learning. We move beyond simple Python scripts to interactive data science tools.
1. Jupyter Notebooks (.ipynb)
• What it is: Unlike standard Python files (.py), Jupyter Notebooks allow you to mix live code, visualizations (graphs), and explanatory text (Markdown) in one document.
• Why we use it: It is the industry standard for communicating data science projects. It allows you to tell a story with your data rather than just writing raw code.
• Setup: You installed the Jupyter extension in VS Code to run these notebooks directly in your editor.
2. The Three "God-Tier" Libraries
You cannot write ML code from scratch every time. We installed three fundamental libraries:
1. NumPy (import numpy as np): The fundamental package for scientific computing. It handles arrays and matrices much faster than standard Python lists. It is essential for linear algebra.
2. Scikit-Learn (import sklearn): The most popular library for Machine Learning. It contains pre-built algorithms (like classification, regression, clustering) and tools for data splitting.
3. Matplotlib (import matplotlib.pyplot as plt): The plotting library. We use this to draw graphs, scatter plots, and visualize our decision boundaries.
