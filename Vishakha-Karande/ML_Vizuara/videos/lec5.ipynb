1. Why Python?
Before we install it, you need to know why we use Python and not C++ or Java for ML.
• Simplicity: It reads like English. It is very easy to write and understand.
• Libraries: Other engineers have already written the hard math code for us (libraries like numpy and scikit-learn which we will use later). We just need to connect them.
• Community: If you get stuck, millions of people are there to help on forums like Stack Overflow.

--------------------------------------------------------------------------------
2. The Concept: Interpreter vs. Editor
This is the most confusing part for beginners. We need to install two different things. Think of this like a Relay Race.
1. The Text Editor (VS Code): This is where you write and manage your files. It’s like a fancy Notepad. It holds the baton (the code).
2. The Interpreter (Python): This is the runner. It takes the code from the Editor and actually runs it on your computer's CPU to get the result.
We need both. The Editor to write, the Interpreter to run.

--------------------------------------------------------------------------------
3. Step-by-Step Installation Guide
Step 1: Install the Interpreter (Python)
1. Go to Google and search "Download Python".
2. Click the official link (python.org) and download the latest version for your OS (Windows or Mac).
3. Run the installer.
    ◦ Tip: If on Windows, make sure to check the box that says "Add Python to PATH" during installation.
4. Verification: Once installed, you technically have Python. You could run code in a black terminal window, but that is hard to manage. That’s why we need Step 2.
Step 2: Install the Editor (VS Code)
There are many editors (PyCharm, Sublime Text), but VS Code is currently the industry standard for beginners and pros alike.
1. Search "Download VS Code" on Google.
2. Download and install it for your OS.
Step 3: Connect Them (The Bridge)
Now you have the Runner (Python) and the Notepad (VS Code), but they don't know each other exists yet.
1. Open VS Code.
2. On the left sidebar, look for the Extensions icon (it looks like four squares).
3. Search for "Python".
4. Click Install on the first result (usually by Microsoft). This teaches VS Code how to understand Python files.
5. Link the Interpreter:
    ◦ Press Command + Shift + P (Mac) or Control + Shift + P (Windows) to open the command palette.
    ◦ Type "Python: Select Interpreter".
    ◦ Click on the "Recommended" option (this should be the Python version you installed in Step 1).
